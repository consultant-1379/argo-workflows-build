```python
import yaml  

with open(FOSSA_DEPENDENCIES_FILE) as f:
    fossa_dependencies = yaml.load(f, Loader=yaml.FullLoader)
```

```python
def validate_src_downloadLink(link):
    if link is None:
        return True
    elif link.endswith(".zip") or link.endswith(".tar.gz") or link.endswith(".tar"):
        return True
    else:
        return False
```

```python
import semver 
def validate_version(version):
    if version is None:
        return True
    else:
        try: 
            semver.VersionInfo.parse(version)
        except ValueError:
            return False
        return True 
```

```python
def validate_source(source):
    if source is None:
        return True
    else:
        if source == 'git':
            return True
        else:
            return False
```

```python
for deps in fossa_dependencies['dependencies']:
    if deps['bazaar']['register'] == 'yes':
        ID = deps['ID']
        isDownloadURLValid = validate_src_downloadLink(deps['bazaar']['src_download_link'])
        isVersionSemver = validate_version(version=deps['bazaar']['component_version'])
        isSourceValid = validate_source(source=deps['additional_info']['fossa-attribution']['Source'])
        print(f"\n\nValidating {ID}...")
        if not isVersionSemver:
            print(f"Version is not valid for {ID}. if its a Commit SHA, modify the version to be a short commit")
            print(f"first 7 characters of the commit SHA is {deps['bazaar']['component_version'][:7]}")
        if not isDownloadURLValid:
            print(f"Download URL is not valid for {ID}. Please check the URL")
        if not isSourceValid:
            print(f"Source is for {ID} is not Git. Please check the source")
        print(f"Validation for {ID} is complete\n\n")
        
        
```
